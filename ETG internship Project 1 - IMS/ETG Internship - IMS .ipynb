# Instructions on running this program.
## I've Excluded While loop where ever necessary since we can run that particular cell how many ever times we want and this is not one whole program( divided by cells)

### This program is designed to :
1. Read data from the Inventory and store it to dictionary variable
2. Add products to the Inventory
3. Delete products from the Inventory
4. Take product ID and Quantity as inputs from the user and generate the bill for that product
5. Update the Inventory after the purchasing
6. Create Sales.JSON file to store the information of products sold
7. Finally Tell us the total amount earned by end of the day and display the products sold with quantities

### INSTRUCTIONS
1. Read data from the record.json file and store it in a dictionary,To do that run the first cell(Below READ FROM INVENTORY Heading)
2. To ADD a new product or quantities to the the existing products, Execute second cell(Below Adding New Item to The Inventory Heading)
3. To DELETE a product Execute Third cell(Below To Delete a Particular product from the inventory Heading)
4. To take input from the user, generate bill for that product, Updating the inventory and to create sales Dictionary run Fourth Cell (Below Purchasing and updating the inventory after purchase Heading)
5. Finally, to crate a sales.json file run the last cell (Below Sales JSON Heading)

### Flow of Running the program (VERY IMPORTANT)  
1. Run Code Cell 2 at the start of the day(only once), since that is initialization step
2. Run Code Cell 3 and 4 when ever necessary during the day
3. Run Code Cell 5 when ever customer comes until the end of the day
4. Run the Code Cell 8 only when your service is closed to create sales.json file of the day and know how much you've earned , i.e at the end of the day (This will be last cell you'll run for the day and only once)

#   Reading from Inventory

```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)

sales = {}
        
count = 0

sum = 0
```

# Adding New Item to The Inventory

```python
Answer = input("Enter N for adding new product E for adding quantities to existing product")
if Answer in ['N','n']: 
    prod_id = input("Enter product ID:")
    if prod_id in record.keys():
        print("This product is already present in the inventory ")
    else:
        prod_name = input("Enter the Product Name:")
        mrp = int(input("Enter the Maximum retail price:"))
        quant = int(input("Enter quantity:"))
        C_mark = input("Enter the color mark(Red or Green) of the product:")
        bestbydate = input("Enter Best by date use of the product in YYYY-MM-DD format:")
        ans = input("Enter is discount is available on the product or not (Yes or No):")
        Amt = int(input("Enter the discount amount:"))
        record[prod_id] = {"Prod_Name": prod_name, 'MRP': mrp, 'Quantities_Avl': quant, "Color_Mark": C_mark, "BestByDate": bestbydate, "Dis_Avl": ans, "Dis_Amt":Amt}
        print("The product has been added successfully")
elif Answer in ['E','e']:
    prod_id = input("Enter product ID:")
    Quantity = input("Enter the amount of quantities to the product")
    record[prod_id]["Quantities_Avl"] += int(Quantity)
    print("The Quantities has been added successfully")
else:
    print("Enter valid Answer")
    
js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

# To Delete a Particular product from the inventory

```python
prod_id = input("Enter product ID to be deleted from the inventory:")
if prod_id in record.keys():
    del record[prod_id]
    print("The product has been deleted from the inventory")
else:
    print("There is no product with the product ID entered in the Inventory")

for Id in record.keys():
    if record[Id]["Quantities_Avl"] == 0:
        del record[Id]

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

# Features of Purchasing Cell

1. Asks user product ID
2. Checks if the ID is Valid, if not displays 'ID not available'
3. Then asks user for quantities
4. Checks if inventory has that many quantities
   - if yes then prints the bill
   - if no displays available quantities and asks the user if they want to buy those
     - if user enters yes then prints the available quantity
     - if no then terminates the program
5. Prints the bill( also upates sales dictionary)
6. Updates the inventory if the product has zero quantity and also to delete the number of records the customer bought

# Purchasing and updating the inventory after purchase
## The cell is repeated thrice to show how the inventory works
### while using it for real, user can run the same cell how many ever times they want
##### This cell has assumed that users enters integer numbers correctly where ever necessary

```python
import time

ui_prod  = str(input("Enter the product_Id of the product: "))
if ui_prod in record.keys():
    ui_quant = int(input("Enter the quantity you want to buy: "))
    if (record[ui_prod]["Quantities_Avl"] - ui_quant) >= 0 :
        count += 1
        print("HELLO! Here's your bill for product number",ui_prod)
        print("___________________________________________________")
        print("Product:", record[ui_prod]['Prod_Name'])
        print("Price:", record[ui_prod]['MRP'])
        print("Quantities you're buying:", ui_quant)
        if record[ui_prod]["Dis_Avl"] == "Yes":
            print("There is Discount Available!\nThe price after discount is:",record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt'])
        if record[ui_prod]["Color_Mark"] == "Green":
            print("This is product is lacto-vegetarian food")
        else:
            print("The color mark is Non-vegetarian food")
        print("Use this product by",record[ui_prod]["BestByDate"],"for better usage")
        print("Billing Amount: ", (record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt']) * ui_quant)
        print('This product was bought on:', time.asctime(time.localtime(time.time()) ))
        if record[ui_prod]['Dis_Avl'] != "No":
            print("You Saved",record[ui_prod]['Dis_Amt'] * ui_quant,"Rupees")
        print('This product was bought on:', time.asctime(time.localtime(time.time()) ))
        print("Have a Good day!")
        print("___________________________________________________")
        record[ui_prod]['Quantities_Avl'] = record[ui_prod]['Quantities_Avl'] - ui_quant
        # Sales JSON
        sales[count] = {"Prod_ID": ui_prod,"Prod_Name": record[ui_prod]["Prod_Name"], "MRP": record[ui_prod]['MRP'], "Quantities_Sold":ui_quant, "Color_Mark":record[ui_prod]["Color_Mark"], "Final_Price":record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt'], "Amount_Earned":(record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt']) * ui_quant}
        
    else:
        print("There are not enough Quantities Available")
        print("The number of quantites Available are:",record[ui_prod]["Quantities_Avl"])
        print("Would you like to buy those?(Yes or No)")
        if input() == "Yes":
            count += 1
            print("HELLO! Here's your bill for product number",ui_prod)
            print("___________________________________________________")
            print("Product:", record[ui_prod]['Prod_Name'])
            print("Price:", record[ui_prod]['MRP']) 
            print("Quantities you're buying:",record[ui_prod]["Quantities_Avl"])
            if record[ui_prod]["Dis_Avl"] == "Yes":
                print("There is Discount Available!\nThe price after discount is:",record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt'])
            if record[ui_prod]["Color_Mark"] == "Green":
                print("This is product is lacto-vegetarian food")
            else:
                print("The color mark is Non-vegetarian food")
            print("Use this product by",record[ui_prod]["BestByDate"],"for better usage")
            print("Billing Amount: ", (record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt']) * record[ui_prod]["Quantities_Avl"])
            if record[ui_prod]["Dis_Avl"] != "No":
                print("You Saved",record[ui_prod]['Dis_Amt'] * record[ui_prod]["Quantities_Avl"],"Rupees")
            print('This product was bought on:', time.asctime(time.localtime(time.time()) ))
            print("Have a Good day!")
            print("___________________________________________________")
            sales[count] = {"Prod_ID": ui_prod, "Prod_Name": record[ui_prod]["Prod_Name"], "MRP": record[ui_prod]['MRP'], "Quantities_Sold":record[ui_prod]["Quantities_Avl"], "Color_Mark":record[ui_prod]["Color_Mark"], "Final_Price":record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt'], "Amount_Earned":(record[ui_prod]['MRP']-record[ui_prod]['Dis_Amt']) * record[ui_prod]["Quantities_Avl"]}
            record[ui_prod]['Quantities_Avl'] = 0
        else:
            print("The product will be available in 2-3 days, see you soon! ")
    
    # Delete the product from the inventory if the quantities available is zero after purchase       
    if record[ui_prod]["Quantities_Avl"] == 0:
        del record[ui_prod]
    # Update the Inventory after purchase
    js = json.dumps(record)
    fd = open("record.json",'w')
    fd.write(js)
    fd.close()
else:
    print("The product with",ui_prod,"as the ID is not available")
```
